# Import Libreries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Data and Create DataFrame

```python
df= pd.read_csv("Amazon Sale Report.csv", encoding= 'unicode_escape')
df
```

```python
df.head()
```

# Clean Dataset

```python
df.shape
```

```python
df.info()
```

```python
# drop unrelated/blank columns 

df.drop(['New', 'PendingS'], axis= 1, inplace= True)
```

```python
df.shape
```

```python
# Check Null Values in Dataset

df.isnull().sum()
```

```python
# Drop Null Values

df.dropna(inplace= True)
```

```python
df.shape
```

```python
df.columns
```

```python
# Change Data type

df['ship-postal-code']= df['ship-postal-code'].astype('int')
```

```python
df['ship-postal-code'].dtype
```

```python
df['Date']= pd.to_datetime(df['Date'], format= 'mixed')
```

```python
# Rename the Column name

df.rename(columns= {'Qty':'Quantity'}, inplace= True)
```

# Exploratory Data Analysis

```python
df.columns
```

## Size

```python
si= sns.countplot(x= 'Size', data= df)

for bars in si.containers:
    si.bar_label(bars)
```

#### Note- From above graph we can see that the most of people buyes M- Size

## Quantity by Size

```python
qua= df.groupby(['Size'], as_index= False)['Quantity'].sum().sort_values(by= 'Quantity', ascending= False)
qua
```

```python
qua_si= sns.barplot(x= 'Size', y= 'Quantity', data= qua)
plt.title('Total Quantity buyes by Size')
plt.show()
```

#### Note- From above graph we can see that the most of Qty buyes M- Size in the sales

## Courier Status

```python
plt.figure(figsize= (10,5))
sns.countplot(x= 'Courier Status', data= df, hue= 'Status')

plt.show()
```

#### Note- From above graph we can see that the most of orders are shipped by the Courier 

## Category

```python
plt.hist(data= df, x= 'Category', bins= 15, edgecolor= 'black')
plt.title('Category most Buyes')
plt.show()
```

#### Note- From above graph we can see that the most of people buyes are T-shirt

## B2B

```python
df['B2B'].value_counts()
```

```python
x= df['B2B'].value_counts().index
y= df['B2B'].value_counts().values
```

```python
plt.figure(figsize=(4,5))
plt.pie(y, labels= x, autopct= '%1.1f%%')
plt.show()
```

#### Note- From above graph we can see that maximum i.e. 99.2% buyers are retailers and 0.8% are B2B buyers

## Category wise Size

```python
x_data= df['Category']
y_data= df['Size']

plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Category wise Size')
plt.show()
```

## Ship-State

```python
shi_st= df.groupby(['ship-state'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

plt.figure(figsize=(15,5))
sns.barplot(x= 'ship-state', y= 'Amount', data=shi_st)
plt.show()
```

#### Note- From above graph we can see that most of the sales are Maharashtra State

## Ship-City

```python
shi_st= df.groupby(['ship-city'], as_index= False)['Amount'].sum().sort_values(by= 'Amount', ascending= False).head(10)

plt.figure(figsize=(15,5))
sns.barplot(x= 'ship-city', y= 'Amount', data=shi_st)
plt.show()
```

#### Note- From above graph we can see that most of the sales in Bengaluru city

## Conclusion

#### The data analysis reveals that the business has a significant customer base in Maharashtra state, mainly serves retailers, fulfil orders through Amazon experiences high demand for T-shirts and sees M-size as the preferred choice among buyers.
